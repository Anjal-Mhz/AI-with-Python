```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv('wine_quality_classification.csv')
data.head(2)
```

```python
X=data.drop(['quality_label'],axis=1)
y=data['quality_label']
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
X_test,X_train,y_test,y_train=train_test_split(X_scaled,y,test_size=0.3,random_state=42)
```

```python
model=SVC(kernel='linear')
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
c=classification_report(y_test,y_pred)
print(c)
```

```python
#rbf

rbf_model=SVC(kernel='rbf')
rbf_model.fit(X_train,y_train)
```

```python
y_pred2=rbf_model.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred2)
print(cm)
accuracy=accuracy_score(y_test,y_pred2)
print(accuracy)
c=classification_report(y_test,y_pred2)
print(c)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
model3=DecisionTreeClassifier()
model3.fit(X_train,y_train)

y_pred3=model3.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred3)
print(cm)
accuracy=accuracy_score(y_test,y_pred3)
print(accuracy)
c=classification_report(y_test,y_pred3)
print(c)
```

```python
model4=RandomForestClassifier()
model4.fit(X_train,y_train)

y_pred4=model4.predict(X_test)
```

```python
cm=confusion_matrix(y_test,y_pred4)
print(cm)
accuracy=accuracy_score(y_test,y_pred4)
print(accuracy)
c=classification_report(y_test,y_pred4)
print(c)
```
