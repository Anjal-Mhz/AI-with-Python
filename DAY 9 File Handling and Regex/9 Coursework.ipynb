```python
import re
```

```python
#matching all words in a string

text="The quick brown fox hi"
pattern= r'\b\w+\b'
match= re.findall(pattern,text)
print(match)
print(pattern)
```

```python
#pattern = r"\b\w+\b"
#r=raw string  
# eg \n= used to Move to next line    AND  r"\n"=prints \n
#\b Word Boundary : this is where tthe words start or end
#\w = Word character (any one letters, numbers, underscore)
#+  = one or more
# \w+ = match one or more letters, numbers and underscore
```

```python
#Finding ID patterns
#Example 123-45-678

p=r"\b\d{3}-\d{2}-\d{3}\b"
#r =raw string
#r"": \ normal
#\d : single digit (number) (0-9): find single digit
#{} : repetition how many times to repeat
#\d{3} : match exactly 3 digits

text="My number is 234-32-183 and 2321-32-3233"
p=r"\b\d{3}-\d{2}-\d{3}\b"
match= re.findall(p,text)
print(match)
```

```python
#9812345678

pattern=r"\b97\d{8}|98\d{8}\b"
text="My number is 9812365890 and 9767565677"
match=re.findall(pattern,text)
print(match)
```

```python
#Extract email form a string
'''
saiman%noob@gmail.com
angelpro@merocoding.com
hello+%cat@edu.np'''

pattern=r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b"

p=r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\b"
'''
r=raw string
""=back slash normal
\b=word boundary start
[]=match any one character form the list
A-Z=uppercase letters
a-z=lowercase letters
0-9:single digit
._%+- = special character
{2,} = must be 2 or more than 2 numbers 
'''

text="This is Saiman. I am noob. You can email me at saiman%noob@gmail.com or saimannoob@merocodingclass.com"
match=re.findall(pattern,text)
match2=re.findall(p,text)
print(match)
print(match2)
```

```python
# Extracting phone numbers from a text file

with open("sample1.txt", "r") as file:
    content = file.read()

pattern = r"\b9\d{9}\b"
phone_numbers = re.findall(pattern, content)
print("Phone Numbers Found:", phone_numbers)
```

```python
# Count the number of times a word appears

word_to_count = "agaiN"
with open("sample1.txt", "r") as file:
    content = file.read()

matches = re.findall(rf"\b{word_to_count}\b", content, re.IGNORECASE)
print(matches)
print(f"'{word_to_count}' found {len(matches)} times.")
```
