```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
data['service']=le.fit_transform(data['service'])
```

```python
data['flag'].value_counts()
```

```python
data['flag']=LabelEncoder().fit_transform(data['flag'])
```

```python
data.head()
```

```python
data.info()
```

```python
X=data.drop('class',axis=1)
y=data['class']
```

```python
from sklearn.preprocessing import StandardScaler

X_scaled=StandardScaler().fit_transform(X)
```

```python
y=y.map({'normal':0,'anomaly':1})
```

```python
y.sample(4)
```

```python
data['class'].value_counts(normalize=True)
```

```python
from sklearn.ensemble import IsolationForest

iso_forest=IsolationForest(contamination=0.4661,random_state=42)
```

```python
data['if_anomaly']=iso_forest.fit_predict(X_scaled)
```

```python
data.sample(5)
```

```python
data['if_anomaly']=data['if_anomaly'].map({1:0,-1:1})
```

```python
data['if_anomaly']
```

```python
from sklearn.metrics import classification_report

print(classification_report(y,data['if_anomaly']))
```

```python
#One class SVM
from sklearn.svm import OneClassSVM
 
oc_svm=OneClassSVM(kernel='rbf',gamma='auto',nu=0.4661)
```

```python
data['oc_svm']=oc_svm.fit_predict(X_scaled)
```

```python
data['oc_svm']
```

```python
data['oc_svm']=data['oc_svm'].map({1:0,-1:1}) 
```

```python
print(classification_report(y,data['oc_svm']))
```
