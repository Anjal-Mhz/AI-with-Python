```python
#Import libaries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
#Load dataset
df=pd.read_csv('movie_genre_classification_final.csv')
```

```python
df=df.drop(['Language','Country','Director','Production_Company','Lead_Actor'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols=['Title','Description','Content_Rating']
print(categorical_cols)
le=LabelEncoder()
for col in categorical_cols:
    df[col]=le.fit_transform(df[col])
```

```python
df['Genre']=le.fit_transform(df['Genre'])
```

```python
df[['Title','Description','Content_Rating']].head(3)
```

```python
#Selecting Features and Target

X=df[['Title','Budget_USD','Content_Rating']]
y=df['Genre']
```

```python
#Spliting data

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#Model Training

model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
#prediction
y_pred=model.predict(X_test)
```

```python
#Evaluation

acc=accuracy_score(y_test,y_pred)
print("Accuracy is ",acc)

cm=confusion_matrix(y_test,y_pred)
print(cm)

c=classification_report(y_test,y_pred)
print(c)
```

```python
#Ensemble Learning

model1=LogisticRegression(max_iter=200)
model2=DecisionTreeClassifier()
model3=RandomForestClassifier()
```

```python
#Create Voting Classifier

from sklearn.ensemble import VotingClassifier

voting_model=VotingClassifier(estimators=[('lr',model1),('dt',model2),('rf',model3)],voting='hard')
```

```python
#Train the ensemble model

voting_model.fit(X_train,y_train)
```

```python
#Predict the ensemble model

y_pred_vt=voting_model.predict(X_test)
```

```python
#Evaluate
acc=accuracy_score(y_test,y_pred_vt)
print(acc)

cm=confusion_matrix(y_test,y_pred_vt)
print(cm)

c=classification_report(y_test,y_pred_vt)
print(c)

#voting model is not good for regressor type data than classifier type data
#reasons are outliers or 
```
