```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('countries.csv')
id=df[df['country'] == 'India']
np=df[df.country == 'Nepal']
jp=df[df.country == 'Japan']
```

```python
plt.plot(np['year'],np['population'],label="Nepal",color='blue',linestyle='-',marker="s")
plt.plot(id['year'],id['population'],color='green',linestyle='--',marker="o",label="India")
plt.plot(jp['year'],jp['population'],color='red',linestyle='-.',marker="x",label="Japan")
plt.title("Population increased per year")
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

string_data_x = ["apple", "banana", "apple", "orange", "banana"]
string_data_y = ["red", "yellow", "red", "green", "blue"]

# Convert strings to numerical categories
df = pd.DataFrame({'x': string_data_x, 'y': string_data_y})
df['x_encoded'] = pd.factorize(df['x'])[0]
df['y_encoded'] = pd.factorize(df['y'])[0]

print(df)

```

```python
# Create the scatter plot
plt.scatter(df['x_encoded'], df['y_encoded'])
plt.xlabel("X Categories (Encoded)")
plt.ylabel("Y Categories (Encoded)")
plt.title("Scatter Plot of Strings")
plt.xticks(ticks=df['x_encoded'].unique(), labels=df['x'].unique())
plt.yticks(ticks=df['y_encoded'].unique(), labels=df['y'].unique())
plt.show()
```
