```python
import matplotlib.pyplot as plt
```

```python
#Line plot
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
plt.show()
```

```python
Month=['Baisakh','Jestha','Asar','Sharawn','Bhadra']
Saving=[2000,1000,3000,1500,2500]

plt.plot(Month,Saving,label="Line Plot")
plt.xlabel("X-axis:Months")
plt.ylabel("Y-axis:Savings")
plt.title("Savings in 5 months")
plt.legend()
plt.show()
```

```python
grade=['A','B','C','D']
frequency=[20,45,50,100]

plt.bar(grade,frequency,color="green")
plt.xlabel("Grade")
plt.ylabel("frequency")
plt.title('Frequency of grades of students')
plt.show()
```

```python
#Scatter plot
height=[5.3,5.7,5.9,6.1,6.3]
weight=[50,55,60,65,68]

plt.scatter(height,weight,color='black')
plt.xlabel('Height (in feet)')
plt.ylabel('Weight (in kg)')
plt.title('Height and Weight')
plt.show()
```

```python
#Histogram

data=[12,12,13,14,15,15,16,16,16,16,17,18,18]
plt.hist(data,bins=7)
plt.xlabel("Data values")
plt.ylabel("Frequency")
plt.title("Histogram value")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

x1 = [1,2,3,4,5]
y2 = [6,7,8,9,10]

plt.plot(x,y,color="red", linestyle='--', marker="x",label='Customized line plot')
plt.plot(x1,y2,color="green",linestyle="-.",marker="o",label="example 2")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Customization")
plt.legend()
plt.show()
```

```python
import pandas as pd

data={'Catagory':['A','B','C','D'], "Frequency":[10,20,25,30]}
df=pd.DataFrame(data)
print(df)
```
